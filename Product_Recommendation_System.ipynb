```python
data=data.loc[data['UnitPrice']>0]
data=data.loc[data['Quantity']>0]
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['CustomerID'],inplace=True)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
customer_item_matrix=data.pivot_table(index='CustomerID', columns='StockCode', values='Quantity', aggfunc='sum')
```

```python
customer_item_matrix.head()
```

```python
customer_item_matrix=customer_item_matrix.applymap(lambda x: 1 if x>0 else 0)
```

```python
customer_item_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_user_sim_matrix=pd.DataFrame(cosine_similarity(customer_item_matrix))
```

```python
user_user_sim_matrix.head()
```

```python
user_user_sim_matrix.shape
```

```python
user_user_sim_matrix.columns=customer_item_matrix.index
```

```python
customer_item_matrix.index
```

```python
user_user_sim_matrix['CustomerID']=customer_item_matrix.index
```

```python
user_user_sim_matrix.set_index('CustomerID',inplace=True)
```

```python
user_user_sim_matrix.head()
```

```python
#finding Most similar customer
user_user_sim_matrix.loc[12358].sort_values(ascending=False)
```

```python
item_recommended_14145=item_bought_12358-item_bought_14145
```

```python
stock_discription_for_14145=data.loc[data['StockCode'].isin(item_recommended_14145)][['StockCode','Description']].drop_duplicates().set_index('StockCode')
```

```python
stock_discription_for_14145
```

```python
def recommend_product_for_customer(customer1):
    customer2=user_user_sim_matrix.loc[customer1].sort_values(ascending=False).reset_index().iloc[1,0]
    item_bought_customer1=set(customer_item_matrix.loc[customer1][customer_item_matrix.loc[customer1]==1].index)
    item_bought_customer2=set(customer_item_matrix.loc[customer2][customer_item_matrix.loc[customer2]==1].index)
    items_recommended=item_bought_customer1-item_bought_customer2
    return customer2,data.loc[data['StockCode'].isin(items_recommended)][['StockCode','Description']].drop_duplicates()
```

```python
customer,recommendations=recommend_product_for_customer(12358)
```

```python
print(customer)
print('*****************************************************************************')
print(recommendations)
```
