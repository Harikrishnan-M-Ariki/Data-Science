```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wine_data=pd.read_csv(r'winequalityN.csv')
```

```python
wine_data.head()
```

```python
wine_data.describe()
```

```python
#Histogram plot
wine_data.hist('alcohol',bins=10)
```

```python
#Using seaborn
import seaborn as sns
sns.set()
```

```python
sns.displot(wine_data['alcohol'],kde=False,color='blue')
plt.show()
```

```python
sns.displot(wine_data['alcohol'],kde=True,color='blue') #kde--approximation plot
plt.show()
```

```python
sns.set_style('dark')
sns.displot(wine_data['alcohol'])
```

```python
#Scatter plot
sns.lmplot(x='alcohol',y='fixed acidity',data=wine_data) #Line represents that alcohol increases fixed acidity decreases
```

```python
sns.lmplot(x='alcohol',y='fixed acidity',data=wine_data,fit_reg=False)
```

```python
sns.lmplot(x='alcohol',y='chlorides',data=wine_data,hue='type',fit_reg=False)
```

```python
wine_data.quantile([0.25,0.75]) #Quantiles
```

```python
sns.boxplot(data=wine_data['alcohol'])
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.boxenplot(data=wine_data,palette='Set3')
plt.show()
```

```python
#Violin plots
sns.violinplot(data=wine_data['alcohol'])
```

```python
import seaborn as sns
plt.figure(figsize=(12,8))
```

```python
sns.countplot(x='quality',data=wine_data,hue='type')
plt.show()
```

```python
sns.set(style='whitegrid')
titanic_dataset=sns.load_dataset('titanic')
```

```python
sns.catplot(x='class',data=titanic_dataset,y='survived',hue='sex',height=6,kind='bar',palette='muted')
```

```python
#Density plot
sns.kdeplot(wine_data.alcohol,wine_data.density)
```

```python
#Joint Distribution plot
sns.jointplot(x='alcohol',y='density',data=wine_data)
```

```python
#Factor plot:
g=sns.factorplot(x='free sulfur dioxide',y='alcohol',data=wine_data,hue='quality',col='quality',kind='point')
```

```python
#empirical cumulative distribution function
x=np.sort(wine_data['alcohol'])
y=np.arange(1,len(x)+1)/len(x)
_=plt.plot(x,y,marker='.',linestyle='none')
plt.margins(0.05)
_=plt.xlabel('Percent of alcohol in wine')
_=plt.ylabel('ECDF')
plt.grid(True)
plt.show()
```
